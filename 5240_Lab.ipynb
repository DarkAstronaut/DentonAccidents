```python
def entropy3V(a,b,c):
  pa = a / (a + b + c)
  pb = b / (a + b + c)
  pc = c / (a + b + c)
  e = -pa * np.log2(pa) - pb * np.log2(pb) - pc * np.log2(pc)
  return round(e,2)
```
