```python
dAccidents['Clearence Time'] = dAccidents['Cleared Time'] - dAccidents['Accident Time']
```

```python
dAccidents['Clearence Time']
```

```python
dAccidents['Clearence Time (Minutes)'] = dAccidents['Clearence Time'].dt.total_seconds()/60
```

```python
round(dAccidents['Clearence Time (Minutes)'].mean(),2)
```

```python
# Dropping $1000 Damage to Any One Person's Property
dAccidents.drop("$1000 Damage to Any One Person's Property", axis=1, inplace=True)
```

```python
dAccidents['Adjusted Average Daily Traffic Amount'].value_counts()
```

```python
# Replacing Sentinel Values - No Data
dAccidents['Adjusted Average Daily Traffic Amount'].replace('No Data', np.nan, inplace=True)
```

```python
dAccidents['Adjusted Average Daily Traffic Amount'] = dAccidents['Adjusted Average Daily Traffic Amount'].astype(float)
```

```python
dAccidents.info()
```

```python
dAccidents['Road Class'].unique()
```

```python
roadClasses = list(dAccidents['Road Class'].unique())
```

```python
#dAccidents[dAccidents['Road Class'] == 'INTERSTATE']['Adjusted Average Daily Traffic Amount'].median()
roadMedianList = []
for road in roadClasses:
  roadMedian = dAccidents[dAccidents['Road Class'] == road]['Adjusted Average Daily Traffic Amount'].astype(float).median()
  roadMedianList.append([road, roadMedian])
```

```python
roadMedianList
```

```python
#dAccidents[(dAccidents['Road Class'] == 'INTERSTATE') & (dAccidents['Adjusted Average Daily Traffic Amount'].isna())]
```

```python
for road in roadMedianList:
    dAccidents.loc[(dAccidents['Road Class'] == road[0]) & (dAccidents['Adjusted Average Daily Traffic Amount'].isna()),'Adjusted Average Daily Traffic Amount'] = road[1]
```

```python
dAccidents['Adjusted Average Daily Traffic Amount'].isna().value_counts()
```

```python
dAccidents['Adjusted Average Daily Traffic Amount'] = dAccidents['Adjusted Average Daily Traffic Amount'].fillna(0)
```

```python
round(dAccidents['Adjusted Average Daily Traffic Amount'].mean(),2)
```

```python
for road in roadClasses:
  c = int(dAccidents['Road Class'][dAccidents['Road Class'] == road].count())
  if c < 10:
    dAccidents.loc[dAccidents['Road Class']== road, 'Road Class'] = 'CITY STREET'
```

```python
for road in roadClasses:
  c = int(dAccidents['Road Class'][dAccidents['Road Class'] == road].count())
  print(road,": ",c)
```

```python
dAccidents['Speed Limit'].value_counts()
```

```python
roadSpeedMedianList
```

```python
dAccidents['Speed Limit'] = dAccidents['Speed Limit'].astype(float)
```
